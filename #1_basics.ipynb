*   We can compare the tensor performance between on CPU and GPU (using colab T4GPU). Generally when training models we will use the GPU because more faster than CPU you can see the result at the below.

```python
import time

device_cpu = "cpu"
device_gpu = "cuda" if torch.cuda.is_available() else "cpu"

# Creating Large Matrix
size = 10000
x_cpu = torch.randn(size, size, device=device_cpu)
x_gpu = torch.randn(size, size, device=device_gpu)

start_time = time.time()
result_cpu = x_cpu @ x_cpu
cpu_time = time.time() - start_time

start_time = time.time()
result_gpu = x_gpu @ x_gpu
gpu_time = time.time() - start_time

print(f"CPU (Intel i7-12700H) execution time: {cpu_time:.4f} seconds")
print(f"GPU (RTX 3050) execution time: {gpu_time:.4f} seconds")

if device_gpu == "cuda":
    speedup = cpu_time / gpu_time
    print(f"GPU is approximately {speedup:.2f} times faster than CPU!")
else:
    print("GPU is not available, tested only on CPU.")
```
